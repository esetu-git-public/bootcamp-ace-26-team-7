# 01 Exploratory Data Analysis

This notebook will explore class counts, distribution of dimensions, color distributions, and plot visual samples of road and bridge cracks.

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from PIL import Image
from src.config import Config
```

## Class Distribution

Count the number of raw images available for each surface defect category.

```python
class_counts = {}
for class_name in Config.CLASSES:
    class_dir = os.path.join(Config.RAW_DATA_DIR, class_name)
    if os.path.exists(class_dir):
        images = [f for f in os.listdir(class_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        class_counts[class_name] = len(images)
    else:
        class_counts[class_name] = 0

print("Class distribution:", class_counts)
```

```python
# Plotting the distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=list(class_counts.keys()), y=list(class_counts.values()), palette="viridis")
plt.title("Class Distribution of Raw Dataset")
plt.ylabel("Number of Images")
plt.xlabel("Defect Category")
plt.show()
```
